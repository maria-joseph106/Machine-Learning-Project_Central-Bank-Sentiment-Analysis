## Unified Payments Interface (UPI): Overview and Sentiment Analysis Relevance

The **Unified Payments Interface (UPI)** is a payment system that was developed by the National Payments Corporation of India (NPCI), an entity regulated by the Reserve Bank of India (RBI). UPI facilitates seamless, instant transactions between any two parties’ bank accounts via mobile devices. It allows combining multiple bank accounts into a single application, supporting both peer-to-peer and person-to-merchant transactions and merging multiple banking features under one application. The RBI launched UPI as a pilot in April 2016.

UPI is now responsible for the vast majority of all digital transactions in India, making it a backbone of the country's payment infrastructure and a catalyst for digital financial inclusion.

## RBI Data

As mentioned in the previous section, the data for the analysis is obtained from the Reserve Bank of India (RBI) and NationalPayments Corporation of India (NPCI) website. Firstly, we collect the RBI data .

This below script automates the process of collecting **all search result entries** from the RBI (Reserve Bank of India) website for the keyword **“upi”**, along with their associated dates. Because the RBI website uses **JavaScript-based pagination (`Next` button + postbacks)**, normal `requests`-based scraping fails. Therefore, we use **Selenium (Headless Chrome)** to behave like a real browser and extract content from multiple pages. There are 744 results for this search as of the date the project was done.



**What the Script Does**
1. Opens the RBI search results in a headless browser, the script launches a headless Chrome browser using Selenium and loads:

https://rbi.org.in/scripts/SearchResults.aspx?search=upi

This allows full rendering of JavaScript and ASP.NET postback-driven pagination.


2. Parses each page using BeautifulSoup

After rendering, we extract:

* **The clickable link** (`a.sub_title2_link`)
* **The date** (`div.sub_title2_detail`)

These items are stored as dictionaries in a Python list.

3. Navigates through all pages

The script looks for the **"Next"** button.
If it exists → Selenium clicks it → next page loads → extraction continues.
If it doesn’t → scraping stops.

4. Stores all results in a Pandas DataFrame

Each record contains:

| Field  | Meaning                                         |
| ------ | ----------------------------------------------- |
| `Link` | Full URL to the RBI page for that search result |
| `Date` | The date displayed for that result              |

The final DataFrame can be saved to a CSV file if required.





```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import time

BASE_URL = "https://rbi.org.in"
SEARCH_URL = f"{BASE_URL}/scripts/SearchResults.aspx?search=upi"

# ---------------------------
# Setup Selenium Headless Chrome
# ---------------------------
options = Options()
options.add_argument("--headless=new")         # Run Chromedriver in headless mode
options.add_argument("--no-sandbox")
options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=options)

# Load the search results page
driver.get(SEARCH_URL)

records = []   # List to store {"Link": ..., "Date": ...}
page = 1

# ---------------------------
# Main scraping loop
# ---------------------------
while True:
    time.sleep(2)   # Allow time for the page to fully render (JS execution)
    
    # Parse HTML with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Each search result entry is contained inside <div id="annual2">
    for div in soup.select("div#annual2"):

        # Extract the hyperlink inside the container
        link_tag = div.select_one("a.sub_title2_link")
        # Extract the date element
        date_tag = div.select_one("div.sub_title2_detail")

        if link_tag:
            # Relative link → Full URL
            href = link_tag.get("href", "").strip()
            full_url = urljoin(BASE_URL, href)

            # Extract date text (if present)
            date_text = date_tag.get_text(strip=True) if date_tag else ""

            # Save into our records list
            records.append({
                "Link": full_url,
                "Date": date_text
            })

    print(f"Page {page} scraped ({len(records)} total links)")

    # ---------------------------
    # Pagination: Move to next page
    # ---------------------------
    next_links = driver.find_elements(By.LINK_TEXT, "Next")

    # If "Next" button doesn't exist → stop scraping
    if not next_links:
        print(" No more pages.")
        break

    # Click "Next" to go to the next page
    next_links[0].click()
    page += 1

# Close the headless browser
driver.quit()

# ---------------------------
# Convert to Pandas DataFrame
# ---------------------------
df = pd.DataFrame(records)

print("\n Preview of scraping :")
print(df.head())


df.to_csv("rbisearchlinks_dates.csv", index=False)
```


This script takes a cleaned DataFrame (`df_clean`) that contains:

* `Link` — the URL of an RBI search-result item
* `Date` — the date associated with that item

It then visits **each page**, scans for **PDF hyperlinks**, and stores the results in a final DataFrame.
To speed things up, it uses **ThreadPoolExecutor** for concurrency and **tqdm** for a progress bar.

---
What the Code Does — Step-by-Step

1. Load cleaned search-result DataFrame.

2. Define a function `fetch_pdfs(row)`

   * Takes one row containing a link + date
   * Downloads the webpage
   * Extracts all `<a>` tags ending in `.pdf`
   * Converts relative URLs to absolute URLs
   * Returns a list of `{pdf_link, date}` dictionaries

3. Run `fetch_pdfs` in parallel using 10 worker threads.

   * Each worker processes a different link
   * `tqdm` shows a progress bar as each task finishes

4. Aggregate results into one DataFrame containing:

   * PDF link
   * Corresponding date






Note that the next step may be computationally intensive and time consuming and in order to skip it, it is possible to go to the next line of code after that directly download the data that is collected in this step. 

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm  # For progress bar

# The cleaned DataFrame containing search results:
# Columns: ["Link", "Date"]
df = df_clean

def fetch_pdfs(row):
    """
    Given one row of the DataFrame (with URL + Date),
    this function downloads the page, finds all PDF links,
    converts relative links into absolute URLs, and returns
    them as a list of dictionaries.
    """

    url = row['Link']
    date = row['Date']
    results = []
    
    try:
        # Fetch HTML content
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        #Parse page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all <a href="...pdf">
        pdf_links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].lower().endswith('.pdf')]
        
        # Convert relative links to absolute links
        for pdf in pdf_links:
            if not pdf.startswith('http'):
                pdf = requests.compat.urljoin(url, pdf)

            results.append({'pdf_link': pdf, 'date': date})


    except requests.exceptions.RequestException as e:
        print(f"Error accessing {url}: {e}")
    
    return results

# -------------------------------
# Run PDF extraction in parallel
# -------------------------------

all_results = []

# Use up to 10 worker threads for speed
with ThreadPoolExecutor(max_workers=10) as executor:
    # Submit all scraping tasks
    futures = {executor.submit(fetch_pdfs, row): idx for idx, row in df.iterrows()}
    
    # Display progress bar as tasks complete
    for future in tqdm(as_completed(futures), total=len(futures), desc="Processing links"):
        all_results.extend(future.result())

# Convert the collected list to a DataFrame
pdf_df = pd.DataFrame(all_results)

print(pdf_df)

# Optionally save:
# pdf_df.to_csv("rbi_pdf_links.csv", index=False)
```

The next part downloads all PDFs listed in a dataframe (pdfs), naming them in a structured way using the date and a counter.
1. A folder /pdf_files is created inside the current working directory to store all PDFs.

2. For each PDF link: we extract the URL, extract the date and maintain a counter for each date to build a unique filename

3. Each file is downloaded using wget.download(). If a download fails: It retries up to 3 times, with a 5-second delay between attempts.
If all retries fail: The index of the failed download is saved in failed_downloads.

4. Instead of stopping the whole script, failures are logged in a list so they can be retried later.

The reason for downloading each file is PDF content isn’t embedded in HTML pages; it exists as a separate binary .pdf file, so you must download it to read it. Attempting to extract text directly from a remote PDF URL is often unreliable because pdfplumber expects a complete file-like binary stream, and chunked downloads or premature HTTPS closures can corrupt the PDF. Downloading locally ensures you can safely reopen the file multiple times, retry extraction if needed, and inspect any problematic PDFs later.

NOTE : The number of failed PDFs detected in the failed_pdfs.txt list was very small. After manually checking each of these links, all of them turned out to be non-functional or invalid URLs (e.g., removed files, redirects without a PDF, or dead links). Therefore, they are not revisited.

```python
import os
import time
import wget
import pandas as pd

# ---------------------------------------
# pdfs = your dataframe containing:
#   - pdf_link : URL of the PDF
#   - date     : date extracted from RBI search results
# ---------------------------------------

# Create folder where PDFs will be saved
download_dir = os.path.join(os.getcwd(), "pdf_files")
os.makedirs(download_dir, exist_ok=True)

# Dictionary to count how many PDFs belong to each date
date_counts = {}

# List to store indices (serial numbers) of failed downloads
failed_downloads = []

# Loop through every row in the dataframe
for i, row in pdfs.iterrows():
    url = row['pdf_link']
    date_str = row['date']

    # Count the number of PDFs for this date
    count = date_counts.get(date_str, 0) + 1
    date_counts[date_str] = count

    # Filename format: DATE_#COUNT.pdf
    filename = f"{date_str}_{count}.pdf"
    filepath = os.path.join(download_dir, filename)

    # Retry logic: up to 3 attempts
    retries = 3
    for attempt in range(retries):
        try:
            wget.download(url, filepath)
            print(f"\nDownloaded {i}: {filename}")
            break  # stop retrying if successful
        except Exception as e:
            print(f"\nFailed {i} attempt {attempt+1}: {e}")
            time.sleep(5)  # wait before retrying

    # This executes only if the 'for attempt' loop never 'breaks'
    else:
        print(f"Skipped {i}: {url}")
        failed_downloads.append(i)  # add row number to list

print("Download process finished!")
print(f"Failed downloads (serial numbers): {failed_downloads}")
```

This entire notebook section does the following:


1. Look into `./pdf_files/` and find all downloaded PDFs.
2. For each PDF:
* extract its text safely using `pdfplumber`
* check if text contains UPI-related keywords
* save extracted text into `./upi_texts/`
* store metadata (date, filename, text)
3. A `ThreadPoolExecutor` is used to speed up extraction by processing 4 PDFs at the same time.

NOTE : The file failed_pdfs.txt contained the indices of PDFs that could not be processed. Upon inspection, it became clear that these entries were not actual PDF documents. They had been incorrectly saved with a .pdf extension despite the files themselves not existing, which caused all processing attempts to fail.



As we have seen before the following code also takes time, the option to skip and download the data is provided in the cell after it.

```python
import os
import re
import glob
import pdfplumber
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed
import pandas as pd

# Folders
pdf_folder = "./pdf_files"         # Folder where all downloaded PDFs are stored
filtered_folder = "./upi_texts"    # Folder where extracted text files will be saved
os.makedirs(filtered_folder, exist_ok=True)

# Keywords to filter PDFs — only PDFs containing these terms will be kept
keywords = ["upi", "unified payments interface"]


# ------------------------------
# Helper Functions
# ------------------------------

def extract_text_from_pdf(file_path):
    """Safely extract raw text from a PDF using pdfplumber."""
    try:
        text = ""
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text + "\n"
        return text
    except Exception as e:
        print(f"Failed to extract text from {file_path}: {e}")
        return None


def contains_keywords(text, keywords):
    """Return True if text contains any UPI-related keyword."""
    text_lower = text.lower()
    return any(kw.lower() in text_lower for kw in keywords)


def process_pdf(pdf_path):
    """
    Extract text, filter, and return metadata.
    """
    filename = os.path.basename(pdf_path)

    # Extract date prefix from filename
    match = re.match(r"(.+?)_(\d+)\.pdf", filename)
    date_str = match.group(1) if match else filename.replace(".pdf", "")

    # Extract PDF text
    text = extract_text_from_pdf(pdf_path)

    if not text or not contains_keywords(text, keywords):
        return None, filename  # failed or irrelevant PDF

    # Return metadata for sentiment analysis
    return {"date": date_str, "filename": filename, "text": text}, None


# ------------------------------
# Parallel Processing
# ------------------------------

pdf_files = sorted(glob.glob(os.path.join(pdf_folder, "*.pdf")))
results = []
failed_pdfs = []

max_workers = 4

print("Submitting PDFs to thread pool...")

# NEW: Inner progress bar (submission)
with tqdm(total=len(pdf_files), desc="Submitting PDFs") as submit_bar:
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        future_to_pdf = {}
        for pdf in pdf_files:
            future = executor.submit(process_pdf, pdf)
            future_to_pdf[future] = pdf
            submit_bar.update(1)

        print("Processing PDFs...")

        # Outer progress bar (completion)
        for future in tqdm(as_completed(future_to_pdf),
                           total=len(pdf_files),
                           desc="Processing PDFs"):

            res, failed = future.result()

            if res:
                results.append(res)
            if failed:
                failed_pdfs.append(failed)


# ------------------------------
# Save Results
# ------------------------------

df_results = pd.DataFrame(results)

with open("failed_pdfs.txt", "w") as f:
    for pdf in failed_pdfs:
        f.write(pdf + "\n")

print(f"Saved failed PDFs list to failed_pdfs.txt")
```

* Re-open the dropdown (prevents stale-element errors)
* Select the year via JS click
* Wait for table refresh
* Scrape all rows of the statistics table

4. We extract the table data and for each row :
* Month
* Banks Live
* Volume (Mn)
* Value (Cr)

All entries are appended to `all_data`.

* Volume (Mn) refers to the number of transactions in millions.

* Value (Cr) refers to the total transacted amount in crores of Indian rupees, where one crore equals ten million (10,000,000) rupees.



```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time



# ----------------- Setup -----------------
driver = webdriver.Chrome()  # Ensure chromedriver is installed and in PATH
url = "https://www.npci.org.in/product/upi/product-statistics"
driver.get(url)

wait = WebDriverWait(driver, 10)

# ----------------- Open the year dropdown -----------------
dropdown_button = wait.until(EC.element_to_be_clickable(
    (By.CSS_SELECTOR, ".dropdown.product-stat-dropdown-black-arrow .dropdown-toggle")
))
# Click using JS to ensure it opens
driver.execute_script("arguments[0].click();", dropdown_button)
time.sleep(1)

# Get all year buttons
year_buttons = driver.find_elements(By.CSS_SELECTOR, ".dropdown-menu.product-stat-dropdown .dropdown-cell")
num_years = len(year_buttons)

# ----------------- Scrape tables -----------------
all_data = []

for i in range(num_years):
    # Re-fetch dropdown and year buttons to avoid stale element issues
    dropdown_button = driver.find_element(By.CSS_SELECTOR, ".dropdown.product-stat-dropdown-black-arrow .dropdown-toggle")
    driver.execute_script("arguments[0].click();", dropdown_button)
    time.sleep(0.5)

    year_buttons = driver.find_elements(By.CSS_SELECTOR, ".dropdown-menu.product-stat-dropdown .dropdown-cell")
    year_button = year_buttons[i]
    year_text = year_button.text

    # Click the year using JS to avoid ElementNotInteractableException
    driver.execute_script("arguments[0].click();", year_button)
    time.sleep(2)  # wait for table to update

    # Scrape the table
    table = driver.find_element(By.CSS_SELECTOR, "table.custom-table")
    rows = table.find_elements(By.CSS_SELECTOR, "tbody tr")

    for row in rows:
        cells = row.find_elements(By.TAG_NAME, "td")
        if len(cells) < 4:  # skip empty or malformed rows
            continue
        month = cells[0].text
        banks_live = cells[1].text
        volume_mn = cells[2].text
        value_cr = cells[3].text
        all_data.append([year_text, month, banks_live, volume_mn, value_cr])

# ----------------- Close driver -----------------
driver.quit()

# ----------------- Convert to DataFrame -----------------
df_upistats = pd.DataFrame(all_data, columns=["Financial Year", "Month", "No. of Banks Live", "Volume (Mn)", "Value (Cr)"])

# Clean numeric columns
df_upistats["No. of Banks Live"] = pd.to_numeric(df_upistats["No. of Banks Live"], errors='coerce')
df_upistats["Volume (Mn)"] = pd.to_numeric(df_upistats["Volume (Mn)"].str.replace(',', ''), errors='coerce')
df_upistats["Value (Cr)"] = pd.to_numeric(df_upistats["Value (Cr)"].str.replace(',', ''), errors='coerce')

# Reset index
df_upistats.reset_index(drop=True, inplace=True)

print(df_upistats)
```

The `get_openai_batch_results_progress` function given below automates **batched sentiment and expectation extraction** using the OpenAI API. Here's how it works step by step:

1. * The input `text_entries` is divided into batches of size `batch_size` (default 5).
   * Batching reduces the number of API calls and is efficient for large datasets.

2. * For each batch, a combined text block is created with numbered entries.
   * A structured system prompt instructs the model to return **sentiment** and **expectations** in a precise, parseable format.

3. * Each batch is sent to the `gpt-4.1-mini` model using `client.responses.create`.
   * If the API call fails, it retries up to `max_retries` times with a `delay` between attempts.

4. * Using a **regular expression**, the function extracts the sentiment score and expectation points for each text entry.
   * If parsing fails, `None` values are recorded to maintain alignment with the input.

5. * Successfully parsed entries are added to the `results` list with `date`, `filename`, `sentiment`, and `expectations`.
   * Entries for which the API or parsing fails after all retries are stored in `failed_entries`.

6. The function returns two lists:

     * `results`: Successfully extracted sentiment and expectations.
     * `failed_entries`: Entries that couldn’t be processed.



```python
def get_openai_batch_results_progress(text_entries, batch_size=5, max_retries=3, delay=2):
    """
    Runs batched sentiment + expectation extraction using the NEW OpenAI API.
    Handles batching, parsing, and retry logic.
    """

    results = []        # stores successful outputs
    failed_entries = [] # tracks failed PDFs / text entries

    # Loop through the data in chunks (batches) of size batch_size
    for i in range(0, len(text_entries), batch_size):

        # Extract a batch (5 by default)
        batch = text_entries[i:i+batch_size]

        # Build a combined text block for the prompt
        batch_text = "\n\n".join([
            f"Text {idx+1}:\n{entry['text']}"
            for idx, entry in enumerate(batch)
        ])

        # System prompt for the model
        prompt = f"""
You are a financial analyst. For each of the following texts about UPI (Unified Payment Interface):

1) Provide a continuous sentiment score from -1 (very negative) to +1 (very positive).
2) Provide two points: likelihood of increased UPI transaction volume, and potential adoption trends.

Respond in the exact format:

Text X:
Sentiment: <score>
Expectations: <point1>; <point2>

{batch_text}
"""

        
        # Retry logic (try up to max_retries times)
        
        success = False
        for attempt in range(max_retries):
            try:
                # Call OpenAI "Responses" API (new format)
                response = client.responses.create(
                    model="gpt-4.1-mini",
                    input=prompt
                )

                # Extract text output
                output = response.output_text.strip()

                # Parse output for each item in the batch
                for idx, entry in enumerate(batch):

                    # Regex pattern for expected format
                    pattern = (
                        rf"Text {idx+1}:\s*"
                        r"Sentiment:\s*([-+]?\d*\.?\d+)\s*"
                        r"Expectations:\s*(.*)"
                    )

                    match = re.search(pattern, output, re.DOTALL)

                    # If parsed successfully
                    if match:
                        sentiment = float(match.group(1))
                        expectations = match.group(2).strip()

                        results.append({
                            "Date": entry["date"],
                            "File name": entry["filename"],
                            "Sentiment": sentiment,
                            "Expectations": expectations
                        })

                    # If parsing fails → store None values
                    else:
                        results.append({
                            "Date": entry["date"],
                            "File name": entry["filename"],
                            "Sentiment": None,
                            "Expectations": None
                        })

                success = True
                break  # exit retry loop when batch succeeds

            except Exception as e:
                # If API fails, retry after delay
                print(f"OpenAI batch failed (attempt {attempt+1}/{max_retries}): {e}")
                time.sleep(delay)

        # If all retries failed, mark batch items as failed
        if not success:
            for entry in batch:
                failed_entries.append(entry["filename"])

    return results, failed_entries
```

### **Random Forests — Foundational Algorithm**

Random Forests are an ensemble learning method that constructs a large number of decision trees and aggregates their predictions to improve generalization performance. Each tree is trained on a different bootstrap sample of the original dataset, introducing **data randomness** through sampling with replacement. Further randomness is injected at the feature level: at every split, only a randomly selected subset of features is considered. This reduces correlation between trees and ensures model diversity.

During training, each decision tree recursively partitions the feature space by selecting the split that maximally reduces impurity — typically measured by **Gini impurity** for classification or **Mean Squared Error (MSE)** for regression. Once all trees are trained, predictions are combined via **majority voting** (classification) or **averaging** (regression).

By aggregating many weakly correlated learners, Random Forests reduce variance while maintaining low bias, making them robust to overfitting, noise, and nonlinear feature interactions. Additionally, they provide measures of **feature importance** based on impurity reduction, aiding interpretability in analytical applications.








### **XGBoost — Foundational Algorithm**

Extreme Gradient Boosting (XGBoost) is a scalable and regularized implementation of gradient boosting that builds an ensemble of decision trees sequentially to minimize a differentiable loss function. Unlike Random Forests, XGBoost builds trees **additively**, where each new tree is trained to correct residual errors from previous trees. At iteration ( t ), the model updates predictions using:

$$
\hat{y}^{(t)} = \hat{y}^{(t-1)} + \eta \cdot h_t(x)$$

where (h_t(x)) is a newly learned tree and ( \eta ) is the learning rate controlling update magnitude. XGBoost optimizes its objective using **second-order gradients**, incorporating both the first derivative (loss gradient) and second derivative (Hessian), resulting in more precise updates and faster convergence.

A key innovation in XGBoost is its explicit **regularization framework** that penalizes model complexity, including the number of leaves and their weights:

$$
\Omega(h_t) = \gamma T + \frac{1}{2} \lambda \sum w_j^2
$$

This reduces overfitting and improves generalization. XGBoost also supports **shrinkage**, **column subsampling**, and **parallelized tree construction**, making it computationally efficient and scalable to large datasets. Its ability to model nonlinear relationships, capture complex feature interactions, and maintain strong predictive accuracy has made it a leading choice in high-performance machine learning applications.

### Value

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import xgboost as xgb

# Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_trainval)
y_pred_rfval = rf.predict(X_test)
print("Random Forest RMSE:", np.sqrt(mean_squared_error(y_testval, y_pred_rfval)))

# XGBoost
xgbr = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
xgbr.fit(X_train, y_trainval)
y_pred_xgbval = xgbr.predict(X_test)
print("XGBoost RMSE:", np.sqrt(mean_squared_error(y_testval, y_pred_xgbval)))
```

```python
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit

# Use TimeSeriesSplit for cross-validation
tscv = TimeSeriesSplit(n_splits=3)

# Random Forest basic grid search
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5]
}
rf_grid = GridSearchCV(
    RandomForestRegressor(random_state=42),
    rf_param_grid,
    cv=tscv,
    scoring='neg_mean_squared_error'
)
rf_grid.fit(X_train, y_trainval)
print("Best RF Params:", rf_grid.best_params_)
print("Best RF RMSE (CV):", np.sqrt(-rf_grid.best_score_))

# XGBoost basic grid search (with TimeSeriesSplit)
xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.03, 0.1, 0.2],
    'max_depth': [3, 6, 10]
}
xgb_grid = GridSearchCV(
    xgb.XGBRegressor(random_state=42),
    xgb_param_grid,
    cv=tscv,
    scoring='neg_mean_squared_error'
)
xgb_grid.fit(X_train, y_trainval)
print("Best XGB Params:", xgb_grid.best_params_)
print("Best XGB RMSE (CV):", np.sqrt(-xgb_grid.best_score_))

```

```python
# Get best estimators
best_rf = rf_grid.best_estimator_
best_xgb = xgb_grid.best_estimator_

best_rf.fit(X_train, y_trainval)
y_pred_rfval_cv = best_rf.predict(X_test)

best_xgb.fit(X_train, y_trainval)
y_pred_xgbval_cv = best_xgb.predict(X_test)
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df_model['Date'].iloc[train_size:], y_testval, label='Actual')
plt.plot(df_model['Date'].iloc[train_size:], y_pred_xgbval, label='Predicted (XGB)')
plt.plot(df_model['Date'].iloc[train_size:], y_pred_xgbval_cv, label='Predicted (XGB) (Best Estimator)')
plt.legend()
plt.title("UPI Growth: Actual vs Predicted")
plt.show()
```
